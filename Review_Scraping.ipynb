```python
import numpy as np
import pandas as pd
```

```python
from google_play_scraper import app # app details
from google_play_scraper import Sort, reviews_all # all reviews
```

## Wysa: Anxiety, therapy chatbot

Link: https://play.google.com/store/apps/details?id=bot.touchkin

ID: bot.touchkin

```python
id1 = 'bot.touchkin'
```

```python
details_wysa = app(id1, lang='en', country='us')
details_wysa
```

```python
# Count/No. of reviews
details_wysa['reviews']
```

```python
'''
sleep_milliseconds=0, # defaults to 0
lang='en', # defaults to 'en'
country='us', # defaults to 'us'
sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT/Sort.NEWEST
count=3, # defaults to 100
filter_score_with=5 # defaults to None(means all score)
'''

all_reviews = reviews_all(
    id1,
    sleep_milliseconds=0, # defaults to 0
    lang='en', # defaults to 'en'
    country='bd', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT
)
#all_reviews
```

```python
reviews = pd.json_normalize(all_reviews)
reviews
#reviews.head()
```

```python
reviews.isnull().sum()
```

```python
reviews['content'].count()
```

```python
reviews.to_csv('wysa_reviews.csv')
```

## Wysa: Mental Health Support

Link: https://apps.apple.com/us/app/wysa-mental-health-support/id1166585565

ID: wysa-mental-health-support

```python
from app_store_scraper import AppStore
from pprint import pprint

wysa_app_reviews = AppStore(country="us", app_name="wysa-mental-health-support")
wysa_app_reviews.review(how_many=100000)

pprint(wysa_app_reviews.reviews)
print(wysa_app_reviews.reviews_count)
print(wysa_app_reviews)
```

```python
reviews = pd.json_normalize(wysa_app_reviews.reviews)
reviews
```

```python
reviews.to_csv('wysa_app_reviews.csv')
```
